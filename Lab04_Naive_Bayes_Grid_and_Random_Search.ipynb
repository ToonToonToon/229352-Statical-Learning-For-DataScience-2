```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

from scipy.stats import uniform
from sklearn.metrics import f1_score
from scipy.stats import uniform
```

```python
train = fetch_20newsgroups(subset='train')
test = fetch_20newsgroups(subset='test')

Xtrain = train.data[:3000]
ytrain = train.target[:3000]
Xtest = test.data[:500]
ytest = test.target[:500]

print("X:", len(Xtest))
print("y:", len(ytest))
```

### Naive Bayes [(Documentation)](https://scikit-learn.org/stable/modules/generated/sklearn.naive_bayes.MultinomialNB.html)

```python
pipeline = Pipeline([('tfidf', TfidfVectorizer(stop_words='english')),
    ('nb', MultinomialNB())])

pipeline.fit(Xtrain, ytrain)

y_pred = pipeline.predict(Xtest)

print(classification_report(ytest, y_pred))
```

### Random Search Cross-Validation [(Documentation)](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.RandomizedSearchCV.html)

### Uniform distribution in `Scipy` [(Documentation)](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.uniform.html)

```python
parameters = {'nb__alpha': uniform(loc=0, scale=10)}

clf = RandomizedSearchCV(pipeline, parameters, n_iter=7)

clf.fit(Xtrain, ytrain)
```

```python
ypred = clf.predict(Xtest)

print(classification_report(ytest, ypred))
```

#### Exercise

1. For the Naive Bayes model, use grid search 5-fold cross-validation across different values of `alpha` to find the best model.

2. For the best value of `alpha`, compute the `f1_macro` score on the test set.
* What value of `alpha` did you obtain?
* What is the model's `f1_macro` score?

3. Repeat Exercise 1 and 2 for **random search** 5-fold cross validation across different values of `alpha`. Compute the `f1_macro` score on the test set.
* What value of `alpha` did you obtain?
* Did you get a better `f1_macro` score compared to grid search in Exercise 2?

```python
params_grid = {'nb__alpha': [0.001, 0.01, 0.1, 0.5, 1.0, 5.0, 10.0]}
```

```python
grid_nb = GridSearchCV(pipeline, params_grid, cv=5, scoring='f1_macro', n_jobs=-1)

grid_nb.fit(Xtrain, ytrain)

print("Best value of alpha:", grid_nb.best_params_['nb__alpha'])
```

```python
ypred_grid = grid_nb.predict(Xtest)
f1_grid = f1_score(ytest, ypred_grid, average='macro')
print(f"Model's f1_macro score: {f1_grid:.4f}")
```

```python
print(classification_report(ytest, ypred_grid))
```

```python
params_dist = {'nb__alpha': uniform(loc=0, scale=5)}
```

```python
rand_nb = RandomizedSearchCV(pipeline, params_dist, cv=5, scoring='f1_macro',
                             n_iter=20, n_jobs=-1, random_state=42)
```

```python
rand_nb.fit(Xtrain, ytrain)

print("\n=== Random Search Results ===")
print("Best value of alpha:", rand_nb.best_params_['nb__alpha'])
```

```python
ypred_rand = rand_nb.predict(Xtest)
f1_rand = f1_score(ytest, ypred_rand, average='macro')
print(f"Model's f1_macro score: {f1_rand:.4f}")
```

```python
print(classification_report(ytest, ypred_rand))
```

```python
print(f"Grid Search F1: {f1_grid:.4f}")
print(f"Random Search F1: {f1_rand:.4f}")

if f1_rand > f1_grid:
    print("Result: Random Search got a BETTER score.")
elif f1_rand < f1_grid:
    print("Result: Grid Search got a BETTER score.")
else:
    print("Result: Both scores are EQUAL.")
```
